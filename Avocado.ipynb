```python
import pandas as pd
import numpy as np

df = pd.read_csv('avocado.csv')
df
```

```python
#Checking for missing values in the DataFrame
df.isnull().sum()
```

```python
# Removing the region and Date features from the DataFrame
df.drop(['region', 'Date'], axis = 1, inplace = True)
df
```

```python
#Scaling the nesscary Data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df.iloc[:, 2:9] = scaler.fit_transform(df.iloc[:, 2:9])
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

#Encodes the Company column
df['type'] = le.fit_transform(df['type'])
df
```

```python
x_columns = 12
X = df.iloc[:, 2:x_columns].values
y = df['AveragePrice'].values
y

from sklearn.model_selection import train_test_split
train_ratio = 0.80
test_ratio = 0.20
validation_ratio = 0.5

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_ratio)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = validation_ratio)
# print(X_train.shape)
# print(X_test.shape)
# print(X_valid.shape)
y
```

```python
k = [2,3,4,15,20]

scores = []
from sklearn.neighbors import KNeighborsRegressor
for i in k:
    k_model = KNeighborsRegressor(n_neighbors = i)
    k_model.fit(X_train, y_train)
    y_pred = k_model.predict(X_test)
    y_pred
    k_score = k_model.score(X_valid, y_valid)
    scores.append(k_score)
    score_k = max(scores)

bestK = scores.index(max(scores))
print(f'The bestscore is {score_k} with k index at {bestK}')
```

```python
from sklearn.metrics import r2_score
model = KNeighborsRegressor(n_neighbors=15)
model.fit(X_train, y_train)
y_pred = model.predict(X_test) 
score = model.score(X_test,y_test)
r2_score=r2_score(y_test,y_pred)
print('r2_score: ',r2_score)
```
